```python
import pandas as pd
import numpy as np
import sqlite3
import click
from contextlib import closing
from zipline.utils.calendars import get_calendar
```

```python
def _fmt_date(dt):
    return pd.to_datetime(dt).strftime('%Y-%m-%d') + " 00:00:00"
```

```python
start_day = '1999-10-21 00:00:00'
end_day = '2000-11-28 00:00:00'
#sids = [118267, 199623]
sids = [118267]
filename = '/home/c.cerbo/.zipline/data/sharadar/latest/prices.sqlite'
fields = ['close']
days_in_sessions = 280
calendar = get_calendar('NYSE')
sessions = calendar.sessions_in_range(start_day, end_day)

raw_arrays = []
with sqlite3.connect(filename) as conn:
    for field in fields:
        query = "SELECT date, sid, %s FROM prices WHERE sid in (%s) and date >= '%s' AND date <= '%s';" \
                % (field, ",".join(map(str, sids)), str(start_day), str(end_day))
        df = pd.read_sql_query(query, conn)
        result = df.pivot(index='date', columns='sid', values=field)
        result = result.reindex(index=list(map(_fmt_date, sessions)), columns=sids)
        raw_arrays.append(result.values)

raw_arrays
```
